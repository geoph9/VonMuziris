```python
import pandas as pd
import numpy as np
```

```python
import os
# Make sure you have downloaded the data files.
assert os.path.isfile("descriptions.xls"), os.listdir()
```

## Data set 1 
(single excel with two sheets)

```python
df1_descs = pd.read_excel("./202 SKUs -Jewellery Data Set.xls", sheet_name="Details").drop_duplicates()
df1_images = pd.read_excel("./202 SKUs -Jewellery Data Set.xls", sheet_name="Images").drop_duplicates()
df1_images.head(2)
```

```python
df1_images.dropna().shape, df1_images.shape
```

```python
df1_descs.rename(columns={'SKU CODE' : 'SKU NAME'}, inplace=True)
df1_descs.drop(columns=["GENDER", "MATERIAL", "Asset Name", "PRODUCT DESCRIPTION"], inplace=True)
df1_descs = df1_descs.applymap(lambda x: x.strip())
df1_descs.head(2)
```

```python
df1 = pd.merge(df1_descs, df1_images, on='SKU NAME', how='left')
print(df1_descs.shape, df1_images.shape, df1.shape)
df1.head()
```

```python
df1['description'] = ['']*len(df1['COLOR'])
df1['images'] = list(map(lambda x: [j for j in x if isinstance(j, str)], df1[['Image name {}'.format(i) for i in range(1, 4)]].values.tolist()))
df1.drop(columns=['Image name {}'.format(i) for i in range(1, 4)], inplace=True)
df1.columns = ['id', 'color', 'name', 'description', 'images']
df1.dropna(inplace=True)
print(df1.shape)
df1.head()
```

## Data set 2
(2 different excel files, one for descriptions and one for images)

```python
df2_descs = pd.read_excel("./descriptions.xls").drop(columns=['QTY', 'SIZE']).drop_duplicates()
df2_images = pd.read_excel("./images.xls").drop_duplicates()
df2_images.head()
```

```python
df2_descs.drop(columns=['TAX CLASS', 'BRAND NAME', 'GENDER', 'MATERIAL', 'TRANSFER PRICE', 'APPROX. FINAL WEIGHT'], inplace=True)
df2_descs.rename(columns={'SKU CODE' : 'SKU NAME'}, inplace=True)
df2_descs = df2_descs.applymap(lambda x: x.strip())
print(df2_descs.shape, df2_images.shape)
df2_descs.head()
```

```python
df2 = pd.merge(df2_descs, df2_images, on='SKU NAME', how='left')
print(df2_descs.shape, df2_images.shape, df2.shape)
df2.head()
```

```python
df2['images'] = list(map(lambda x: [j for j in x if isinstance(j, str)], df2[['Image name {}'.format(i) for i in range(1, 4)]].values.tolist()))
df2.drop(columns=['Image name {}'.format(i) for i in range(1, 4)], inplace=True)
df2.columns = ['id', 'color', 'name', 'description', 'images']
df2.dropna(inplace=True)
print(df2.shape)
df2.head()
```

## Combine datasets into one

```python
df = df1.append(df2).drop_duplicates(['id', 'color', 'name', 'description'])
print(df.shape)
df.head()
```
